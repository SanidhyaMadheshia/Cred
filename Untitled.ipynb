```python
# Cell — Descriptive stats + IQR-based outlier handling (pandas-version safe)

# 1) Pick numeric columns explicitly (works on all pandas versions)
num_cols = df.select_dtypes(include=[np.number]).columns
if len(num_cols) == 0:
    raise ValueError("No numeric columns found to summarize / filter.")

# Descriptive statistics for numeric columns only
desc = df[num_cols].describe().T
print("Numeric descriptives (first 10 rows):")
print(desc.head(10))

# 2) IQR bounds per numeric column
q1 = df[num_cols].quantile(0.25)
q3 = df[num_cols].quantile(0.75)
iqr = q3 - q1
lower = q1 - 1.5 * iqr
upper = q3 + 1.5 * iqr

# 3) Row filter: keep rows within bounds for ALL numeric columns
within_bounds = ((df[num_cols] >= lower) & (df[num_cols] <= upper)).all(axis=1)
df_no_outliers = df[within_bounds].copy()

print(f"\nOriginal rows: {len(df)}")
print(f"Rows after IQR filter: {len(df_no_outliers)}")
print(f"Rows dropped: {len(df) - len(df_no_outliers)}")

# 4) Winsorized copy (clip extremes to bounds), written per-column for compatibility
df_winsor = df.copy()
for c in num_cols:
    df_winsor[c] = df_winsor[c].clip(lower=lower[c], upper=upper[c])

# 5) Compare class distributions before/after (Target preferred, else RiskBucket, else Rating)
label_col = "Target" if "Target" in df.columns else ("RiskBucket" if "RiskBucket" in df.columns else "Rating")
before = df[label_col].value_counts()
after_filter = df_no_outliers[label_col].value_counts().reindex(before.index, fill_value=0)
after_winsor = df_winsor[label_col].value_counts().reindex(before.index, fill_value=0)

compare = (
    pd.DataFrame({
        "before": before,
        "after_filter": after_filter,
        "after_winsor": after_winsor
    })
    .assign(dropped_by_filter=lambda t: t["before"] - t["after_filter"])
    .sort_index()
)

print(f"\nClass distribution comparison by '{label_col}':")
print(compare)
```

```python
# Cell — Split numeric/categorical, correlations, multicollinearity diagnostics

# 1) Split features
numericals  = df.select_dtypes(include=[np.number]).copy()
categorical = df.select_dtypes(include=["object", "string"]).copy()

print(f"Numeric cols: {len(numericals.columns)}  |  Categorical cols: {len(categorical.columns)}")
print("First 10 numeric cols:", list(numericals.columns[:10]))
print("First 10 categorical cols:", list(categorical.columns[:10]))

# 2) Pearson correlations (numeric only)
corr = numericals.corr()                 # works across pandas versions
# Mask upper triangle for a cleaner plot
mask = np.triu(np.ones_like(corr, dtype=bool), k=1)

# Dynamic figure size (kept reasonable for large matrices)
n = len(corr.columns)
side = min(1 + 0.5*n, 18)
plt.figure(figsize=(side, side))
sns.heatmap(corr, mask=mask, annot=False, vmin=-1, vmax=1, center=0, square=True,
            cbar_kws={"shrink": 0.8})
plt.title("Feature Correlations (Pearson)")
plt.tight_layout()
plt.show()

# 3) List the most correlated pairs (absolute corr >= 0.90)
thr = 0.90
pairs = (
    corr.where(mask)                  # keep only upper triangle pairs
        .stack()
        .rename("corr")
        .reindex()                    # ensure a Series
        .reset_index()
        .rename(columns={"level_0":"feat_a","level_1":"feat_b"})
        .assign(abs_corr=lambda t: t["corr"].abs())
        .sort_values("abs_corr", ascending=False)
)
high_corr = pairs[pairs["abs_corr"] >= thr]
print(f"Highly correlated pairs (|r| >= {thr}):")
print(high_corr.to_string(index=False) if not high_corr.empty else "None above threshold.")

# 4) Quick VIF (Variance Inflation Factor) — no extra libraries
#    VIF_i = 1 / (1 - R_i^2) from regressing feature i on the others
def compute_vif(X):
    X0 = X.dropna()
    # standardize (numerically stable)
    Xs = (X0 - X0.mean()) / X0.std(ddof=0)
    vifs = {}
    cols = Xs.columns
    for col in cols:
        y = Xs[col].values
        Xo = Xs.drop(columns=[col]).values
        # add intercept
        X_design = np.column_stack([np.ones(Xo.shape[0]), Xo])
        beta, *_ = np.linalg.lstsq(X_design, y, rcond=None)
        y_hat = X_design @ beta
        ss_res = np.sum((y - y_hat)**2)
        ss_tot = np.sum((y - y.mean())**2)
        r2 = 0.0 if ss_tot == 0 else (1 - ss_res/ss_tot)
        vifs[col] = np.inf if (1 - r2) == 0 else 1.0 / (1 - r2)
    return pd.Series(vifs).sort_values(ascending=False)

# To keep it fast, you can limit to candidate model features if you wish.
# Here we run over all numeric features:
vif_series = compute_vif(numericals)
print("\nTop 15 VIFs (higher => more multicollinearity):")
print(vif_series.head(15).round(2))
```

```python
# Cell — Feature selection (SelectFromModel on numeric block) + define X,y per maker

# 1) Target for feature selection (maker uses letter Rating)
y_fs = data["Rating"].astype(str)

# Align X (numeric block after high-corr drop) and y, handling NaNs safely
X_fs = numerical.replace([np.inf, -np.inf], np.nan)
keep_idx = X_fs.dropna().index
X_fs = X_fs.loc[keep_idx]
y_fs = y_fs.loc[keep_idx]

# 2) SelectFromModel with GradientBoosting (default threshold = mean importance)
SFM = SelectFromModel(estimator=GradientBoostingClassifier(random_state=42))
SFM.fit(X_fs, y_fs)

feature_mask = SFM.get_support()
feature_names = X_fs.columns[feature_mask].tolist()
print("Selected numeric features by GradientBoosting (SFM):")
print(feature_names)

# (Optional) show importances sorted, to understand what got picked
importances = getattr(SFM.estimator_, "feature_importances_", None)
if importances is not None:
    imp_tbl = (
        pd.DataFrame({"feature": X_fs.columns, "importance": importances})
          .sort_values("importance", ascending=False)
          .reset_index(drop=True)
    )
    print("\nTop importances from GradientBoosting:")
    print(imp_tbl.head(15).to_string(index=False))

# 3) Define predictors and target exactly as the maker specified
y = data["Rating"].astype(str)

X = data[[
    "Symbol",
    "Year",
    "Sector",
    "netProfitMargin", "returnOnAssets", "debtRatio",
    "freeCashFlowOperatingCashFlowRatio", "freeCashFlowPerShare",
    "operatingCashFlowSalesRatio"
]].copy()

print("\nX dtypes:")
print(X.dtypes)
print("\nX shape:", X.shape, " | y shape:", y.shape)
```

```python
# Cell — Train/test split, label-encode y, scale numeric features (maker-aligned)

# 1) Target + predictors
label_col = "Target" if "Target" in data.columns else "Rating"
y = data[label_col].astype(str)

X = data[[
    "Symbol", "Year", "Sector",
    "netProfitMargin", "returnOnAssets", "debtRatio",
    "freeCashFlowOperatingCashFlowRatio", "freeCashFlowPerShare",
    "operatingCashFlowSalesRatio"
]].copy()

# 2) Split (stratified for balanced class proportions)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42, stratify=y
)

# 3) Label-encode targets (maker used LabelEncoder)
encoder = LabelEncoder()
y_train_encoded = encoder.fit_transform(y_train)
y_test_encoded  = encoder.transform(y_test)

print("Mapping:", dict(zip(encoder.classes_, range(len(encoder.classes_)))))
print("X_train shape:", X_train.shape)
print("y_train shape:", y_train.shape)
print("y_train counts:\n", y_train.value_counts())

# 4) Scale numerics only (fit on train, apply to test)
numericals_train = X_train.select_dtypes(include=[np.number]).copy()
numericals_test  = X_test.select_dtypes(include=[np.number]).copy()

transformer = StandardScaler().fit(numericals_train)
numericals_train_standardized = pd.DataFrame(
    transformer.transform(numericals_train),
    columns=numericals_train.columns, index=numericals_train.index
)
numericals_test_standardized = pd.DataFrame(
    transformer.transform(numericals_test),
    columns=numericals_test.columns, index=numericals_test.index
)

# (Optional: peek)
numericals_train.shape
```

```python
# Version-safe OneHotEncoder + build final matrices (no re-imports)
from sklearn.preprocessing import OneHotEncoder

# if these aren't defined (e.g., after restart), grab them again
if 'categoricals_train' not in globals():
    categoricals_train = X_train.select_dtypes(include=["object","string"]).copy()
    categoricals_test  = X_test.select_dtypes(include=["object","string"]).copy()

# 1) Instantiate OHE with a compatibility shim
try:
    # scikit-learn ≥ 1.2 prefers sparse_output
    ohe = OneHotEncoder(handle_unknown="ignore", drop="first", sparse_output=False)
except TypeError:
    # older versions use 'sparse'
    ohe = OneHotEncoder(handle_unknown="ignore", drop="first", sparse=False)

# 2) Fit on TRAIN only
ohe.fit(categoricals_train)

# 3) Transform and ensure dense arrays
Xt_train_cat = ohe.transform(categoricals_train)
Xt_test_cat  = ohe.transform(categoricals_test)
if hasattr(Xt_train_cat, "toarray"):
    Xt_train_cat = Xt_train_cat.toarray()
    Xt_test_cat  = Xt_test_cat.toarray()

# 4) Build DataFrames with matching index
ohe_cols = ohe.get_feature_names_out(categoricals_train.columns)
categoricals_train_encoded = pd.DataFrame(Xt_train_cat, columns=ohe_cols, index=categoricals_train.index)
categoricals_test_encoded  = pd.DataFrame(Xt_test_cat,  columns=ohe_cols, index=categoricals_test.index)

# 5) Concatenate with scaled numerics
X_train_final = pd.concat([numericals_train_standardized, categoricals_train_encoded], axis=1)
X_test_final  = pd.concat([numericals_test_standardized,  categoricals_test_encoded],  axis=1)

print("X_train_final shape:", X_train_final.shape)
print("X_test_final  shape:", X_test_final.shape)

# keep for saving later
ohe_encoder = ohe
```

```python
# Cell — Concatenate (alias) + SMOTE balancing (maker-aligned)

# 1) Alias the already-built matrices (avoids double concat bugs)
if 'X_train_final' in globals() and 'X_test_final' in globals():
    X_train_encoded = X_train_final.copy()
    X_test_encoded  = X_test_final.copy()
else:
    # Fallback to maker’s concat if previous cell wasn’t run
    numericals_train_standardized.reset_index(drop=True, inplace=True)
    categoricals_train_encoded.reset_index(drop=True, inplace=True)
    X_train_encoded = pd.concat((numericals_train_standardized, categoricals_train_encoded), axis=1)

    numericals_test_standardized.reset_index(drop=True, inplace=True)
    categoricals_test_encoded.reset_index(drop=True, inplace=True)
    X_test_encoded = pd.concat((numericals_test_standardized, categoricals_test_encoded), axis=1)

print("X_train_encoded shape:", X_train_encoded.shape)
print("X_test_encoded  shape:", X_test_encoded.shape)

# 2) Balance training set with SMOTE (train only)
sm = SMOTE(random_state=100, k_neighbors=3)
X_train_SMOTE, y_train_SMOTE = sm.fit_resample(X_train_encoded, y_train_encoded)

# 3) Show class counts after SMOTE
vals, cnts = np.unique(y_train_SMOTE, return_counts=True)
print("Balanced classes (label_id: count):", dict(zip(vals, cnts)))
```

```python
# Cell — Logistic Regression hyperparameter tuning (solver-safe) + pick best

from sklearn.model_selection import RandomizedSearchCV

# Train/val data: use the SMOTE-balanced train set to match earlier training
Xtr, ytr = X_train_SMOTE, y_train_SMOTE
Xte, yte = X_test_encoded, y_test_encoded

# --- Search A: L2-only (lbfgs/newton-cg) ---
space_l2 = {
    "solver": ["lbfgs", "newton-cg"],
    "penalty": ["l2"],
    "multi_class": ["multinomial"],
    "C": [0.001, 0.01, 0.1, 1, 10, 100],
    "fit_intercept": [True, False],
    "max_iter": [500, 1000, 2000],
    "class_weight": ["balanced"],
}

rs_l2 = RandomizedSearchCV(
    LogisticRegression(),
    param_distributions=space_l2,
    n_iter=20,
    cv=5,
    scoring="accuracy",
    n_jobs=1,           # Windows-safe
    random_state=42,
    verbose=0
)
rs_l2.fit(Xtr, ytr)
best_l2 = rs_l2.best_estimator_
acc_l2_te = metrics.accuracy_score(yte, best_l2.predict(Xte))
print("Best L2-only (lbfgs/newton-cg):", rs_l2.best_params_, "| Test acc:", round(acc_l2_te, 4))

# --- Search B: SAGA (l1/l2/elasticnet) ---
space_saga = {
    "solver": ["saga"],
    "penalty": ["l1", "l2", "elasticnet"],
    "multi_class": ["multinomial"],
    "C": [0.001, 0.01, 0.1, 1, 10, 100],
    "fit_intercept": [True, False],
    "max_iter": [500, 1000, 2000],
    "l1_ratio": [0.0, 0.5, 1.0],  # used only when penalty='elasticnet'; ignored otherwise
    "class_weight": ["balanced"],
}
rs_saga = RandomizedSearchCV(
    LogisticRegression(),
    param_distributions=space_saga,
    n_iter=30,
    cv=5,
    scoring="accuracy",
    n_jobs=1,           # Windows-safe
    random_state=42,
    verbose=0
)
rs_saga.fit(Xtr, ytr)
best_saga = rs_saga.best_estimator_
acc_saga_te = metrics.accuracy_score(yte, best_saga.predict(Xte))
print("Best SAGA:", rs_saga.best_params_, "| Test acc:", round(acc_saga_te, 4))

# --- Pick the winner ---
if acc_saga_te >= acc_l2_te:
    LR_tuned = best_saga
    Accuracy_LR_tuned = acc_saga_te
    source = "SAGA"
else:
    LR_tuned = best_l2
    Accuracy_LR_tuned = acc_l2_te
    source = "L2-only"

print(f"\nSelected tuned LR ({source}) — Test Accuracy: {Accuracy_LR_tuned:.4f}")

# Optional: confusion matrix for the tuned LR
LR_tuned_cm = confusion_matrix(yte, LR_tuned.predict(Xte), labels=np.arange(len(encoder.classes_)))
plt.figure(figsize=(8,6))
sns.heatmap(LR_tuned_cm, annot=True, fmt='d', cmap='Blues', cbar=False,
            xticklabels=list(encoder.classes_), yticklabels=list(encoder.classes_))
plt.xlabel('Predicted'); plt.ylabel('Actual'); plt.title('Tuned Logistic Regression — Confusion Matrix')
plt.tight_layout(); plt.show()
```

```python
# Clean + refit best Logistic Regression (no duplicate kwargs)

# 1) Get best params (same as before)
if 'LR_tuned' in globals():
    best_params = LR_tuned.get_params()
else:
    candidates = []
    if 'rs_l2' in globals():   candidates.append(rs_l2)
    if 'rs_saga' in globals(): candidates.append(rs_saga)
    if not candidates:
        raise RuntimeError("No tuned LR found. Run the tuning cell first.")
    winner = max(candidates, key=lambda s: s.best_score_)
    best_params = winner.best_params_

# 2) Sanitize incompatible / duplicate keys
bp = dict(best_params)              # copy
bp.pop("random_state", None)        # avoid multiple values for random_state
if bp.get("penalty") != "elasticnet":
    bp.pop("l1_ratio", None)        # only valid with elasticnet
# If solver is lbfgs or newton-cg, force compatible settings
if bp.get("solver") in ("lbfgs", "newton-cg"):
    bp["penalty"] = "l2"
    bp["multi_class"] = "multinomial"

print("Refitting LR with params:", bp)

# 3) Train on SMOTE-balanced train; evaluate on untouched test
Xtr = X_train_SMOTE if 'X_train_SMOTE' in globals() else X_train_encoded
ytr = y_train_SMOTE if 'y_train_SMOTE' in globals() else y_train_encoded
Xte, yte = X_test_encoded, y_test_encoded

best_LR_model = LogisticRegression(random_state=1234, **bp)
best_LR_model.fit(Xtr, ytr)

y_pred_LR_best = best_LR_model.predict(Xte)
Accuracy_LR_best = metrics.accuracy_score(yte, y_pred_LR_best)
print("Best LR Accuracy:", round(Accuracy_LR_best, 4))

# 4) Confusion matrix with correct label names
labels_ids = np.arange(len(encoder.classes_))
LR_best_cm = confusion_matrix(yte, y_pred_LR_best, labels=labels_ids)
label_names = list(encoder.classes_)

plt.figure(figsize=(8,6))
sns.heatmap(LR_best_cm, annot=True, fmt='d', cmap='Blues', cbar=False,
            xticklabels=label_names, yticklabels=label_names)
plt.xlabel('Predicted'); plt.ylabel('Actual'); plt.title('Best Logistic Regression — Confusion Matrix')
plt.tight_layout(); plt.show()
```

```python
import numpy as np
import pandas as pd
import pickle
import warnings
import shap
import json

# --- Suppress potential warnings ---
warnings.filterwarnings("ignore", category=UserWarning)

# --- Load artifacts ---
try:
    with open("model_LR.pkl", "rb") as f:
        model = pickle.load(f)
    with open("transformer.pkl", "rb") as f:
        scaler = pickle.load(f)
    with open("encoder.pkl", "rb") as f:
        ohe = pickle.load(f)
except FileNotFoundError as e:
    print(f"Error loading model files: {e}")
    exit()

# --- Feature Configuration ---
label_names = ["High Risk", "Highest Risk", "Low Risk", "Medium Risk"]
NUM_COLS = [
    "Year",
    "netProfitMargin", "returnOnAssets", "debtRatio",
    "freeCashFlowOperatingCashFlowRatio", "freeCashFlowPerShare",
    "operatingCashFlowSalesRatio",
]
CAT_COLS = ["Symbol", "Sector"]

def get_feature_names():
    num_features = NUM_COLS
    cat_features = ohe.get_feature_names_out(CAT_COLS)
    return list(num_features) + list(cat_features)

# --- Load background dataset for SHAP ---
df_bg = pd.read_csv("validation_data.csv")

# Prepare numeric & categorical background
X_num_bg = df_bg[NUM_COLS].replace([np.inf, -np.inf], np.nan).fillna(df_bg[NUM_COLS].median(numeric_only=True))
X_cat_bg = df_bg[CAT_COLS].astype("string").copy()

Xn_bg = scaler.transform(X_num_bg)
Xc_bg = ohe.transform(X_cat_bg)
if hasattr(Xc_bg, "toarray"):
    Xc_bg = Xc_bg.toarray()

X_background = np.hstack([Xn_bg, Xc_bg])

# --- Create SHAP explainer once using background dataset ---
explainer = shap.Explainer(model, X_background, feature_names=get_feature_names())

# --- Prediction function ---
def predict_one(manual):
    row = pd.DataFrame([manual])

    X_num = row[NUM_COLS].replace([np.inf, -np.inf], np.nan).fillna(row[NUM_COLS].median(numeric_only=True))
    X_cat = row[CAT_COLS].astype("string").copy()

    Xn = scaler.transform(X_num)
    Xc = ohe.transform(X_cat)
    if hasattr(Xc, "toarray"):
        Xc = Xc.toarray()
    X_final = np.hstack([Xn, Xc])

    # Prediction
    y_id = int(model.predict(X_final)[0])
    label = label_names[y_id] if 0 <= y_id < len(label_names) else f"Unknown ID: {y_id}"
    proba = model.predict_proba(X_final)[0] if hasattr(model, "predict_proba") else None

    # SHAP explanation
    shap_values = explainer(X_final)
    feature_contribs = shap_values.values[0, :, y_id]

    # Top 3 features
    importance = sorted(
        zip(get_feature_names(), feature_contribs),
        key=lambda x: abs(x[1]),
        reverse=True
    )[:3]

    # Build JSON output
    out = {
        "Company": manual["Symbol"],
        "Sector": manual["Sector"],
        "PredictedRating": label,
        "Probabilities": {},
        "TopFeatures": []
    }
    if proba is not None:
        sorted_labels = sorted(label_names)
        for cls, p in zip(sorted_labels, proba):
            out["Probabilities"][cls] = round(float(p), 4)
    for feat, val in importance:
        out["TopFeatures"].append({"Feature": feat, "Contribution": round(float(val), 4)})

    return out

# --- Sample input ---
sample_data = dict(
    Symbol="AAPL",
    Sector="Technology",
    Year=2024,
    netProfitMargin=0.22,
    returnOnAssets=0.18,
    debtRatio=0.56,
    freeCashFlowOperatingCashFlowRatio=0.65,
    freeCashFlowPerShare=8.1,
    operatingCashFlowSalesRatio=0.33
)

results = [predict_one(sample_data)]
print(json.dumps(results, indent=4))
bhai ye code hai shap vale ka
iska output ye hai
[
    {
        "Company": "AAPL",
        "Sector": "Technology",
        "PredictedRating": "Low Risk",
        "Probabilities": {
            "High Risk": 0.0003,
            "Highest Risk": 0.0001,
            "Low Risk": 0.9981,
            "Medium Risk": 0.0015
        },
        "TopFeatures": [
            {
                "Feature": "freeCashFlowPerShare",
                "Contribution": 116585.086
            },
            {
                "Feature": "netProfitMargin",
                "Contribution": -7.3438
            },
            {
                "Feature": "Symbol_AAPL",
                "Contribution": 2.9156
            }
        ]
    }
]
```

```python
# === PART 2: FETCH APPLE NEWS (7D) + VADER SENTIMENT ===

import os, math, re, requests, datetime as dt
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

NEWS_API_KEY = os.getenv("NEWS_API_KEY", "8a7653142e2f47f291ebcc0da8fd219d")  # <-- set this!
NEG_KEYWORDS = [
    "default","bankruptcy","insolvency","liquidation",
    "downgrade","credit watch","going concern",
    "fraud","restatement","restructuring","mass layoff",
    "covenant breach","distress","delisting"
]
analyzer = SentimentIntensityAnalyzer()

def _utcnow():
    return dt.datetime.now(dt.timezone.utc)  # aware UTC (no deprecation)

def _parse_iso_utc(s: str, fallback: dt.datetime):
    """Parse ISO8601; ensure tz-aware UTC."""
    try:
        t = dt.datetime.fromisoformat(s.replace("Z", "+00:00"))
        if t.tzinfo is None:
            t = t.replace(tzinfo=dt.timezone.utc)
        return t.astimezone(dt.timezone.utc)
    except Exception:
        return fallback

def fetch_news_metrics(symbol: str, company_name: str = None, days: int = 7, language="en"):
    """
    Return dict: sentiment_7d ∈ [-1,1], n_articles_7d, neg_event_spike, articles (top 10).
    """
    if NEWS_API_KEY.startswith("PUT_"):
        return {"sentiment_7d": 0.0, "n_articles_7d": 0, "neg_event_spike": False, "articles": []}

    q = f'{symbol}' + (f' OR "{company_name}"' if company_name else "")
    to_dt = _utcnow()                                   # aware
    from_dt = to_dt - dt.timedelta(days=days)

    url = "https://newsapi.org/v2/everything"
    params = {
        "q": q,
        "from": from_dt.date().isoformat(),
        "to": to_dt.date().isoformat(),
        "language": language,
        "sortBy": "publishedAt",
        "pageSize": 100,
        "apiKey": NEWS_API_KEY,
    }
    data = requests.get(url, params=params, timeout=20).json()
    articles = data.get("articles", []) or []

    vals, weights, kept = [], [], []
    neg_spike = False

    for a in articles:
        title = a.get("title") or ""
        desc  = a.get("description") or ""
        text  = (title + ". " + desc).strip()
        if not text:
            continue

        comp = analyzer.polarity_scores(text)["compound"]  # [-1, +1]

        published = a.get("publishedAt") or ""
        t = _parse_iso_utc(published, fallback=to_dt)      # aware
        delta_days = max(0.0, (to_dt - t).total_seconds() / 86400.0)
        w = math.exp(-delta_days / 7.0)

        vals.append(comp * w); weights.append(w)
        kept.append({"publishedAt": published, "title": title, "compound": comp})

        tl = text.lower()
        if any(kw in tl for kw in NEG_KEYWORDS) and comp <= 0:
            neg_spike = True

    sentiment_7d = (sum(vals)/sum(weights)) if weights and sum(weights) > 0 else 0.0
    sentiment_7d = max(-1.0, min(1.0, float(sentiment_7d)))

    news = {
        "sentiment_7d": sentiment_7d,
        "n_articles_7d": len(kept),
        "neg_event_spike": bool(neg_spike),
        "articles": kept[:10],
    }
    print(f"Articles processed: {news['n_articles_7d']}")
    print(f"Average sentiment (7d): {news['sentiment_7d']:.5f}",
          "POSITIVE 📈" if news['sentiment_7d']>0 else "NEGATIVE 📉" if news['sentiment_7d']<0 else "NEUTRAL ➖")
    return news

# Re-run for Apple
apple_news = fetch_news_metrics(symbol="AAPL", company_name="Apple Inc")
apple_news
```

```python
# === PART 3: PREDICT (FUNDAMENTALS) + APPLY NEWS TILT → FINAL OUTPUT ===
import numpy as np
import pandas as pd

LABELS = ["High Risk","Highest Risk","Low Risk","Medium Risk"]  # model's training order

def apply_news_tilt(base_probs, sentiment_7d, neg_event_spike=False, n_articles_7d=None,
                    min_articles=3, alpha=0.20, gamma=1.20, spike_boost=0.25):
    """Small, explainable tilt toward safer/riskier classes based on news; no retrain."""
    base_probs = np.asarray(base_probs, dtype=float)
    if n_articles_7d is not None and n_articles_7d < min_articles:
        return base_probs, {"skipped": "too_few_articles"}

    adj = float(sentiment_7d) - (spike_boost if neg_event_spike else 0.0)
    adj = max(-1.0, min(1.0, adj))

    # class order: [High, Highest, Low, Medium]
    tilt = np.array([
        np.exp(-gamma * adj),   # High
        np.exp(-gamma * adj),   # Highest
        np.exp(+gamma * adj),   # Low
        np.exp(+0.5*gamma*adj)  # Medium (gentler)
    ], dtype=float)

    tilted = base_probs * tilt
    tilted /= tilted.sum()

    final = (1 - alpha) * base_probs + alpha * tilted
    final /= final.sum()

    dbg = {"news_adj": adj, "alpha": alpha, "gamma": gamma,
           "neg_event_spike": bool(neg_event_spike), "n_articles_7d": n_articles_7d}
    return final, dbg

def proba_dict(model, probs):
    """Return a dict of probabilities in the model's class order."""
    out = {}
    for class_id, p in zip(getattr(model, "classes_", range(len(LABELS))), probs):
        cls = LABELS[class_id] if 0 <= class_id < len(LABELS) else str(class_id)
        out[f"proba_{cls.replace(' ','_')}"] = float(p)
    return out

# 1) Fundamentals-only prediction for Apple
row = pd.DataFrame([sample])
X_num = row[NUM_COLS].copy().replace([np.inf, -np.inf], np.nan).fillna(row[NUM_COLS].median(numeric_only=True))
X_cat = row[CAT_COLS].astype("string").fillna("Unknown")

Xn = scaler.transform(X_num)
Xc = ohe.transform(X_cat)
if hasattr(Xc, "toarray"): Xc = Xc.toarray()
X_final = np.hstack([Xn, Xc])

base_probs = model.predict_proba(X_final)[0]
base_label = LABELS[int(np.argmax(base_probs))]

# 2) Apply news overlay using PART 2 metrics
final_probs, dbg = apply_news_tilt(
    base_probs=base_probs,
    sentiment_7d=apple_news["sentiment_7d"],
    neg_event_spike=apple_news["neg_event_spike"],
    n_articles_7d=apple_news["n_articles_7d"],
    min_articles=3, alpha=0.20, gamma=1.20, spike_boost=0.25
)
final_label = LABELS[int(np.argmax(final_probs))]

# 3) Return compact JSON-like output
output = {
    "symbol": sample["Symbol"],
    "sector": sample["Sector"],
    "base": {
        "predicted_rating": base_label,
        "probabilities": proba_dict(model, base_probs)
    },
    "news": {
        "sentiment_7d": apple_news["sentiment_7d"],
        "neg_event_spike": apple_news["neg_event_spike"],
        "n_articles_7d": apple_news["n_articles_7d"],
        # "headlines": apple_news["articles"]  # uncomment to include top headlines
    },
    "after_news": {
        "predicted_rating": final_label,
        "probabilities": proba_dict(model, final_probs),
        "debug": dbg
    }
}
output
```
